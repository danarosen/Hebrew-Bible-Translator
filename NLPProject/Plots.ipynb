```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# example data
np.random.seed(42)
y = np.random.uniform(0,1, 50)
x = np.arange(0, 50,1)
d = np.random.normal(0,1, 10000)

np.random.seed(73)
y1 = np.random.uniform(0,1, 50)

print(y)
print(y1)

print(x)
print(d)
```

```python
# with sns.plotting_context(font_scale=1.5): # so text adjusts automatically  #sns.set_context('paper', font_scale=1.4)
fig, ax = plt.subplots(figsize=(11,8.5)) # size of plots


sns.distplot(d) # just put your data in 


plt.yticks(rotation=0, size=16); #Rotate and size
plt.xticks(rotation=0, size=16);

plt.title("Title", size=20);

plt.ylabel("Density",  size=18)
plt.xlabel("BLEU of Validation",  size=18)

plt.grid() 

# plt.xticks(np.arange(0,100,5)*2, size=1); # if want to use a different set of labels
# plt.xlim([-2, 2])# Limit of the dist


# plt.savefig("example.png", dpi=1000, bbox_inches='tight')
```

```python
fig, ax = plt.subplots(figsize=(11,8.5)) # size of plots


sns.lineplot(ax=ax, y=y, x=x, label="Training", linewidth=3) # just put your data in 
sns.lineplot(ax=ax, y=y1, x=x, label="Validation", linewidth=3) # just put your data in 


plt.yticks(rotation=0, size=16); #Rotate and size
plt.xticks(rotation=0, size=16);

plt.title("Title", size=20);

plt.ylabel("Loss Name",  size=18)
plt.xlabel("Epoch",  size=18)
plt.legend( prop={'size': 16})
plt.grid() 
```

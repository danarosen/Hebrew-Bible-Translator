```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

from google.colab import drive
drive.mount('/content/gdrive', force_remount=True)
```

```python
RUN_NAME = "training"
LOG_FILE_NAME = 'log_hugface_model.csv'
DRIVE_PATH = "/content/gdrive/MyDrive/"
LOG_FILE_FULL = DRIVE_PATH + LOG_FILE_NAME
```

```python
df = pd.read_excel(LOG_FILE_FULL, header=None)
```

```python
df
```

```python
df_data = df.iloc[[22, 24, 26, 27, 29, 30,  31,32, 33, 34]]

df_train = df_data[df_data[1] == "train"]
df_val = df_data[df_data[1] == "val"]
```

```python
fig, ax = plt.subplots(figsize=(11,8.5))
for i in range(len(df_val.iloc[:, 5:])):
  sns.distplot(df_val.iloc[i, 5:].dropna(), label = "Epoch {}".format(i+1), ax=ax, kde_kws={'clip': (0.0, 1.0), "linewidth":3}, bins=50, hist_kws=dict(alpha=.33))


plt.yticks(rotation=0, size=16); #Rotate and size
plt.xticks(rotation=0, size=16);
plt.ylabel("Density",  size=18)
plt.xlabel("Validation BLEU",  size=18)
plt.legend(prop={'size': 16})
plt.xlim([0,1])
plt.show()
```

```python
N = len(df_train.iloc[:, 5:])

fig, ax = plt.subplots(1, N, figsize=(11*N,8.5))
palette = iter(sns.color_palette("colorblind"))

linestyles = iter(list({'-': '_draw_solid', '--': '_draw_dashed', '-.': '_draw_dash_dot', ':': '_draw_dotted', 'None': '_draw_nothing', ' ': '_draw_nothing'}.keys())) #"linestyle":next(linestyles)

for i in range(N):
  sns.distplot(df_train.iloc[i, 5:].dropna(), label = "Epoch {}".format(i+1), ax=ax[i], kde_kws={'clip': (0.0, 20), "linewidth":3, }, bins=50, hist_kws=dict(alpha=.33),  color=next(palette))
  ax[i].set_ylabel("Density",  size=18)
  ax[i].set_xlabel("Epoch {}: Training Loss".format(i+1),  size=18)
  ax[i].set_xlim([0,5])
  for tick in ax[i].xaxis.get_major_ticks():
      tick.label.set_fontsize(16) 
  for tick in ax[i].yaxis.get_major_ticks():
      tick.label.set_fontsize(16) 
      
# plt.yticks(rotation=0, size=16); #Rotate and size
# plt.xticks(rotation=0, size=16);
# plt.legend(prop={'size': 16})


plt.show()
```

```python
N = len(df_val.iloc[:, 5:])

fig, ax = plt.subplots(1, N, figsize=(11*N,8.5))
palette = iter(sns.color_palette("colorblind"))

linestyles = iter(list({'-': '_draw_solid', '--': '_draw_dashed', '-.': '_draw_dash_dot', ':': '_draw_dotted', 'None': '_draw_nothing', ' ': '_draw_nothing'}.keys())) #"linestyle":next(linestyles)

for i in range(N):
  sns.distplot(df_val.iloc[i, 5:].dropna(), label = "Epoch {}".format(i+1), ax=ax[i], kde_kws={'clip': (0.0, 1.0), "linewidth":3, }, bins=50, hist_kws=dict(alpha=.33),  color=next(palette))
  ax[i].set_ylabel("Density",  size=18)
  ax[i].set_xlabel("Epoch {}: Val BLEU".format(i+1),  size=18)
  ax[i].set_xlim([0,1])
  for tick in ax[i].xaxis.get_major_ticks():
      tick.label.set_fontsize(16) 
  for tick in ax[i].yaxis.get_major_ticks():
      tick.label.set_fontsize(16) 
  d= df_val.iloc[i, 5:].dropna()
  ax[i].axvline(d.mean(), color="red", linewidth=3, linestyle=":")
  geo_mean = np.exp(np.log(d.prod(axis=0))/d.notna().sum(0))
  ax[i].axvline(geo_mean, color="red", linewidth=3, linestyle="-")
  # print(geo_mean)

# plt.yticks(rotation=0, size=16); #Rotate and size
# plt.xticks(rotation=0, size=16);
# plt.legend(prop={'size': 16})


plt.show()
```
